# Line Plot

```python
# Mendeklarasikan variabel years yang berisi tahun

years=df_london['year'].unique()
```

```python
# Mendeklarasikan variabel sumsandyears yang berisi tahun yang dijumlahkan

sumsandyear=df_london.groupby('year').sum()
```

```python
# Mendeklarasikan variabel sums berisi nilai value yang dikonversi ke numpy array

sums=sumsandyear['value'].to_numpy()
```

```python
# Membuat fungsi perulanagan untuk mencetak tahun 2008 - 2016
n = len(years)

for i in range(n):
    for j in range(0, n-i-1):
        if years[j] > years[j+1] :
                years[j], years[j+1] = years[j+1], years[j]
                sums[j], sums[j+1] = sums[j+1], sums[j]
 
print("Tahun :  ")
print(years)
print("Jumlah kejadian kriminal yang terjadi di setiap tahun :")
print(sums)
```

```python
print("Grafik garis yang menunjukkan jumlah total kejahatan dalam tahun 2008 to 2016")

crime = np.arange(len(years))

plt.xticks(crime,years)
plt.plot(sums)
plt.xticks(crime,years)
plt.title("Kejadian Kriminal pada tahun 2008 sampai 2016 di London", color='red')
plt.xlabel('Tahun', color='red')
plt.ylabel("Banyak Kasus", color='red')
plt.xticks(rotation=-45 )
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xticks(color='red')
plt.yticks(color='red')
plt.show()
```

# Word Clouds

- Mengimplementasikan Word Clouds Menggunakan 'Word Clouds' Library Dengan Data Yang Relevan Dan Sesuai Dengan Kegunaan Plot/Grafik

```python
# install wordcloud
# !conda install -c conda-forge wordcloud --yes

!pip install wordcloud

# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS
from PIL import Image # converting images into arrays

print('WordCloud is installed and imported')
```

```python
# open the file and read it into a variable london12000
london12000 = open('london12000.txt', 'r').read()

print('File downloaded and saved!')
```

```python
# menggunakan set untuk menghapus stopword yang berlebihan

stopwords = set(STOPWORDS)
```

```python
# instantiate a word cloud object
london_wc = WordCloud(
background_color='white',
max_words=2000, # mengambil 2000 kata pertama saja
stopwords=stopwords)

# generate the word cloud
london_wc.generate(london12000)
```

```python
# memvisualisasikan word cloud

plt.imshow(london_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# mengubah ukuran cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# memvisualisasikan word cloud
plt.imshow(london_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# save mask to london_mask
london_mask = np.array(Image.open('london_mask.png'))

print('Image downloaded and saved')
```

```python
# mengubah ukuran cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# memvisualisasikan word cloud
plt.imshow(london_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# instantiate a word cloud object
london_wc = WordCloud(
background_color='white',
max_words=2000, # mengambil 2000 kata pertama saja
mask=london_mask,
stopwords=stopwords)

# generate the word cloud
london_wc.generate(london12000)

# mengubah ukuran cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# memvisualisasikan word cloud
plt.imshow(london_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Informasi
- Kata Theft, Motor Vehicle, Criminal Damage, dan Burglary merupakan kata yang paling sering muncul dalam 2000 kata pertana

# Choropleth Maps

- Mengimplementasikan London Maps Menggunakan Choropleth Maps

```python
import folium # Mengimpor folium

df_b=df_london[['borough','value']]
world_geo = r'https://raw.githubusercontent.com/radoi90/housequest-data/master/london_boroughs.geojson' 
world_map=folium.Map(location=[51.5072,-0.1275],zoom_start=10,tiles = 'Stamen Terrain')
```

```python
# Mencetak choropleth

world_map.choropleth(
    geo_data=world_geo,
    name='choropleth',
    data=df_b,
    columns=['borough', 'value'],
    key_on='feature.properties.name',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Tingkat kejadian kriminal di Wilayah London'
    )

world_map.save('LondonCrime.html') # menyimpan hasil dalam bentuk gambar ke direktori
display(world_map) # menampilkan map
```
